```python
# 4.1
df = pd.read_csv("data.csv")
# используем LabelEncoder
from sklearn.preprocessing import LabelEncoder
df["target"] = LabelEncoder().fit_transform(df["target"])
print(df.head(), '\n\n Значения колонки target:',
      df.target.unique())
```

```python
# используем OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoder.fit(df[["target"]])
encoded_data = encoder.transform(df[["target"]])
encoded_data_df = pd.DataFrame(encoded_data.toarray(), columns=['iris-setosa', 'iris-versicolor', 'iris-virginica'])
df = df.drop(['target'], axis=1)
df = pd.concat([df, encoded_data_df], axis=1)
print(df.head())

# df = pd.get_dummies(df, columns=["target"], dtype=int)
# print(df.head())
```

```python
# 4.2
def print_save_plots(data, title, columns, bins, file_name=None):
    
    fig, axs = plt.subplots(2, 4, figsize=(28, 6))
    fig.suptitle(title, fontsize=19, fontweight='bold')
    
    for i, column in enumerate(columns):
        sns.histplot(data=data, x=column, bins=bins[i], ax=axs[i // 4][i % 4], palette='tab10', multiple='layer', element = 'step', kde = True)

    if file_name:
        fig.savefig(file_name)
        
    plt.show()

df_np = df.values
columns = df.columns
bins = [10, 20, 20, 20, 20, 20, 20, 20]
print(df_np[:5][:], '...')
```

```python
# 4.3
# Для массива NumPy из п. 4.2 примените StandardScaler, MinMaxScaler, MaxAbsScaler и RobustScaler. 
# Для каждого из результатов постройте гистограммы по каждому признаку без разделения по классам
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, RobustScaler
import matplotlib.pyplot as plt

data_ss = pd.DataFrame(StandardScaler().fit_transform(df_np), columns = columns)
# plt.data_ss.hist(bins=20)
print_save_plots(data_ss, 'Histogram of distribution of StandardScaler data features', columns, bins, 'StandardScaler.png')
```

```python
data_mm = pd.DataFrame(MinMaxScaler().fit_transform(df_np), columns = columns)
# print(data_mm.hist(bins=20))
print_save_plots(data_mm, 'Histogram of distribution of MinMaxScaler data features', columns, bins, 'MinMaxScaler.png')
```

```python
data_ma = pd.DataFrame(MaxAbsScaler().fit_transform(df_np), columns = columns)
# print(data_ma.hist(bins=20))
print_save_plots(data_ma, 'Histogram of distribution of MaxAbsScaler data features', columns, bins, 'MaxAbsScaler.png')
```

```python
robust_rs = RobustScaler(quantile_range = (30,60))
data_rs = pd.DataFrame(robust_rs.fit_transform(df_np), columns = columns)
# print(data_rs.hist(bins=20))
print_save_plots(data_rs, 'Histogram of distribution of RobustScaler data features', columns, bins, 'RobustScaler.png')
```

```python
# 4.4
# реализуем StandardScaler с использованием NumPy
class MyStandardScaler:
    def __init__(self):
        self.mean = None
        self.std = None

    def fit(self, X):
        self.mean = np.mean(X, axis=0)
        self.std = np.std(X, axis=0)
        return self

    def transform(self, X):
        return (X - self.mean) / self.std

    def fit_transform(self, X):
        self.fit(X)
        return self.transform(X)

# проверим на корректную работу
scaler = MyStandardScaler()
scaler.fit(df)
print("Mean:", scaler.mean)
print("Standard deviation:", scaler.std)

data_ss_new = pd.DataFrame(MyStandardScaler().fit_transform(df_np), columns = columns)
# print(data_ss_new.hist(bins=20))
# print(data_ss.hist(bins=20))
print_save_plots(data_ss_new, 'Histogram of distribution of MyStandardScaler data features', columns, bins, 'MyStandardScaler.png')
print_save_plots(data_ss, 'Histogram of distribution of StandardScaler data features', columns, bins)
```

```python
# 4.4
# реализуем MinMaxScaler с использованием NumPy
class MyMinMaxScaler:
    def __init__(self):
        self.min = None
        self.max = None

    def fit(self, X):
        self.min = np.min(X, axis=0)
        self.max = np.max(X, axis=0)
        return self

    def transform(self, X):
        return (X - self.min) / (self.max-self.min)

    def fit_transform(self, X):
        self.fit(X)
        return self.transform(X)

# проверим на корректную работу
scaler = MyMinMaxScaler()
scaler.fit(df)
print("Min:", scaler.min)
print("Max:", scaler.max)

data_ss_new = pd.DataFrame(MyMinMaxScaler().fit_transform(df_np), columns=columns)
# print(data_ss_new.hist(bins=20))
# print(data_ss.hist(bins=20))
print_save_plots(data_ss_new, 'Histogram of distribution of MyMinMaxScaler data features', columns, bins, 'MyMinMaxScaler.png')
print_save_plots(data_ss, 'Histogram of distribution of MinMaxScaler data features', columns, bins)
```

```python
# минимальное, максимальное, среднее значение и дисперсию, после преобразования.
min_ss = round(np.min(data_ss_new, axis=0), 9)
max_ss = round(np.max(data_ss_new, axis=0), 9)
mean_ss = round(np.mean(data_ss_new, axis=0), 9)
std_ss = round(np.std(data_ss_new, axis=0)**2, 9)

table = PrettyTable()
head = list(df.columns.values)
table.field_names = head
table.add_rows(
    [
        min_ss,
        max_ss,
        mean_ss,
        std_ss
    ]
)
table.add_column("Characteristics", ['min', 'max', 'mean', 'std'])

print(table)
```
